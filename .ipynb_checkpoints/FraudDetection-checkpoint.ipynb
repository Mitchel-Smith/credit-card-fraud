```python
from sklearn.model_selection import train_test_split
X = df_credit.iloc[:,:-1]
T = df_credit['Class']
X_train, X_test, t_train, t_test = train_test_split(X,T, test_size = .2, train_size = .8)
```

```python
#Taken from Lab 6
# TODO: import the necessary modules
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

"""
   evaluation report
       create a classification evaluation using
          accuracy, MCC, classification_report, confusion matrix, and ROC Curve (AUC as well)
          
       parameters
       -----------
       model      trained classifier instance
       data       tuple of (X_train, X_test, t_train, t_test)
"""
def evaluation_report(model, data):
    print("{}\n===========================\n".format(model.__class__.__name__))
    X_train, X_test, t_train, t_test = data
    y_test = model.predict(X_test)
    
    #### 
    # TODO Print train & test accuracy
    train_score = model.score(X_train, t_train)
    test_score = model.score(X_test, t_test)# TODO 3.D
    print("Train Accuracy: {}, Test Accuracy: {}".format(train_score, test_score))
    
    
    # TODO  print MCC score
    print("MCC: {}".format(matthews_corrcoef(y_test, t_test)))
    
    # TODO  print classification report
    print("Classification Report: {}".format(classification_report(y_test, t_test)))
    
    # TODO plot confusion matrix
    cm = confusion_matrix(y_test, t_test, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
    disp.plot()
    
    plt.title(model.__class__.__name__)
```

```python
#Let's figure out what size tree we should be building
from sklearn.ensemble import RandomForestClassifier

#build the tree, should use warm start and max depth due to dataset size
rf = RandomForestClassifier(warm_start = True, oob_score = True, max_depth=5)

# Range of `n_estimators` values to explore, keeping it pretty small, only 5 steps
#since the dataset is so large

min_estimators = 50
max_estimators = 250
rng = range(min_estimators, max_estimators, 50)
#Inititalize some arrays to save data in
errors = {'train': [], 'test': [], 'oob': []}

for n_est in rng:
    #for each 50 trees
    rf.set_params(n_estimators=n_est)
    #fit to that
    rf.fit(X_train, t_train)
    #Check the errors
    train_error = 1 - rf.score(X_train, t_train)
    test_error = 1 - rf.score(X_test, t_test)
    oob_error = 1 - rf.oob_score_
    #add them to our arrays
    errors['train'].append(train_error)
    errors['test'].append(test_error)
    errors['oob'].append(oob_error)
#Now plot all of the cells    
plt.plot(rng, errors['train'])
plt.plot(rng, errors['test'])
plt.plot(rng, errors['oob'])

#Let's make the graph pretty
plt.xlim(min_estimators, max_estimators)
plt.legend(['train','test','oob'])
plt.xlabel("# estimators")
```

```python
#It's important to check and make sure that we aren't just prediction all false, so let's run our
#evaluation report and really focus on the CM

#Be sure to set n_estimators based on the previous
#For our case it looks like 100 should be ok
rf = RandomForestClassifier(n_estimators=100, warm_start=True, max_depth=5)
rf.fit(X_train, t_train)
evaluation_report(rf, [X_train, X_test, t_train, t_test])
```

Here we actually did quite well, we caught almost all of the actual frauds. Most of our errors came
from improperly predicting actaully purchases as fraud. For our case, that's not a huge deal because
it's not like the bank is going to go yoink the item from the customer. All a false positive will do
is trigger a phone call or a push alert or something that will take the customer a minute or two to 
clear up. We would love to get the number or false negatives down to 0, but 5/83= 6%

Ok, time for logistic regression. Let's just start off by giving it a run with the default settings
and see where that gets us

```python
from sklearn.linear_model import LogisticRegression
logClf = LogisticRegression(random_state=0)
logClf.fit(X_train, t_train)
evaluation_report(logClf, [X_train, X_test, t_train, t_test])
```

Ok, well that's not too bad for a first go. Let's see if we can get it to be better by including
some additional arguments to our logistic classifier. We know we have an imbalanced dataset, let's
try weighting our parameters. We can also increase the max_iters a bit since we see it hasn't converged yet from the above try. 

```python
logClf = LogisticRegression(random_state=0, class_weight='auto', max_iter = 1000)
logClf.fit(X_train, t_train)
evaluation_report(logClf, [X_train, X_test, t_train, t_test])
```
